## 1. Modeling the Pizza

```
pizzas = [
          ['small', [26, 0], 4.80, 'circle'],
          ['large', [30, 0], 5.50, 'circle'],
          ['party', [46, 33], 13.00, 'rectangle'],
          ['family', [60, 40], 20.00, 'rectangle'],
          ['deal', [32, ], 5.00, 'circle']
]
```

```
pizzas[0]
```

## 2. Determining the Areas

```
import math
```

```
# A function to calculate the area for a given pizza

def pizza_area(size, shape):
  if shape == 'circle':
    area = math.pi * (size[0] / 2) ** 2
  if shape == 'rectangle':
    area = size[0] * size[1]

  return round(area)
```

```
pizza_area([26, 0], 'circle')
```

```
# Calculating the areas of all pizzas

for pizza in pizzas:
  area = pizza_area(pizza[1], pizza[3])
  print(f'Pizza {pizza[0]} has an area of {area}.')
```

**Optional:** Similarly to the PTER function, we can construct a version that takes a pizza (object) as input instead of size and shape.

```
# An alternative function to calculate the area for a given pizza

def pizza_area_2(pizza):
  if pizza[3] == 'circle':
    area = math.pi * (pizza[1][0] / 2) ** 2
  if pizza[3] == 'rectangle':
    area = pizza[1][0] * pizza[1][1]

  return round(area)
```

```
pizza_area_2(pizzas[0])
```

## 3. Determining PTERs and the Best/Worst Pizza

```
# A function to calculate the PTER for a given pizza

def pizza_pter(area, price):
  pter = area / price

  return round(pter, 2)
```

```
# An alternative function to calculate the PTER for a given pizza based on our
# data structure

def pizza_pter_2(pizza):
  area = pizza_area(pizza[1], pizza[3])
  pter = area / pizza[2]

  return round(pter, 2)
```

```
pizza_pter(531, 4.80)
```

```
pizza_pter_2(['small', [26, 0], 4.80, 'circle'])
```

```
pizza_pter_2(pizzas[0])
```

```
# Calculate PTERs of all pizzas

for pizza in pizzas:
  print(f'The PTER for {pizza[0]} is {pizza_pter_2(pizza)}.')
```

```
best_pizza = ''
worst_pizza = ''
best_pter = 0
worst_pter = 9999

for pizza in pizzas:
  #area = pizza_area(pizza[1], pizza[3])
  #pter = pizza_pter(area, pizza[2])

  pter = pizza_pter_2(pizza)

  if pter > best_pter:
    best_pter = pter
    best_pizza = pizza[0]

  if pter < worst_pter:
    worst_pter = pter
    worst_pizza = pizza[0]

print(f'The best pizza is "{best_pizza}" with a PTER of {best_pter}.')
print(f'The worst pizza is "{worst_pizza}" with a PTER of {worst_pter}.')
```
